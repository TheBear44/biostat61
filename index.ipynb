# สัตว์ในสยาม (กลุ่มที่ 2)
## สัตว์พาหะนำโรคและมาตรการจัดการ (ป้องกันและกำจัด) ของร้านค้าประเภทต่าง ๆ ในพื้นที่สยามสแควร์

นำเข้าและเตรียมข้อมูล

```python
import pandas as pd
import numpy as np
df=pd.read_csv('data/data.csv')
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Location'] = df["Location"].astype('category')
df['Type'] = df["Type"].astype('category')
df['ลักษณะของร้าน'] = df["ลักษณะของร้าน"].astype('category')
df['Restroom']=df['Restroom'].map({"ไม่มี": False, "มี": True})
df['Garbage']=df['Garbage'].astype('category')
df['Drain']=df['Drain'].map({"ไม่มี": False, "มี": True})
animals = df["ท่านเคยพบสัตว์ประเภทใดในบริเวณร้านบ้าง"].str.split(", ", expand=True).stack().unique()
df = pd.concat([df, df["ท่านเคยพบสัตว์ประเภทใดในบริเวณร้านบ้าง"].str.get_dummies(", ")], axis=1)
```

### ตารางแสดงจำนวนร้านค้าประเภทต่างๆในการสำรวจครั้งนี้

```python
df['Type'].value_counts()
```

### ตารางแสดงสถานที่ที่พบสัตว์ชนิดต่างๆ

```python
pd.pivot_table(df, values=animals, index=["Location"], aggfunc=np.sum)
```

### ตารางแสดงสถานที่ที่พบสัตว์ชนิดต่างๆ

```python
pd.pivot_table(df, values=animals, index=["Type"], aggfunc=np.sum)
```

### ตารางแสดงชนิดที่รองรับขยะมูลฝอยกับประเภทร้านค้า

```python
pd.crosstab(df["Type"], df["Garbage"])
```

### ตารางแสดงการมีฝาปิดท่อระบายน้ำกับประเภทร้านค้า

```python
pd.crosstab(df["Type"], df["Drain"])
```
