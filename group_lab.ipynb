b) Formulate the hypothesis and statistics test for this association.

$H_{0}$ : There is no difference in cancer death between smoker and non-smoker group.

$H_{A}$ : There is a difference in cancer death between smoker and non-smoker group.

Statistics test: z-test and chi-square test

```python
from statsmodels.stats.proportion import proportions_ztest as ztest
from statsmodels.stats.proportion import proportions_chisquare as chisquare
count = [140, 39]
nobs = [579, 972]
zstat, zpval = ztest(count, nobs)
print("z-test stat: " + "{0:0.3f}".format(zstat) + " p-value: " + '{0:0.3f}'.format(zpval))

chistat, chipval, table = chisquare(count, nobs)
print("chi-square test stat: " + "{0:0.3f}".format(chistat) + " p-value: " + '{0:0.3f}'.format(chipval))
```

There is a statistically significant difference in cancer death between smoker and non-smoker group. ($p<0.01$)

6\. Research has suggested a high rate of alcoholism among patients with primary unipolar depression. An investigation by Winokur and Coryell further explored this possible relationship. In 210 families of females with primary unipolar major depression, they found that alcoholism was present in 89. Of 299 control families, alcoholism was present in 94. Do these data provide sufficient evidence for us to conclude that alcoholism is more likely to be present in families of subjects with unipolar depression? Let $\alpha$ = .05.

```python
from statsmodels.stats.proportion import proportions_ztest as ztest
from statsmodels.stats.proportion import proportions_chisquare as chisquare
count = [89, 94]
nobs = [210, 299]
zstat, zpval = ztest(count, nobs)
print("z-test stat: " + "{0:0.3f}".format(zstat) + " p-value: " + '{0:0.3f}'.format(zpval))

chistat, chipval, table = chisquare(count, nobs)
print("chi-square test stat: " + "{0:0.3f}".format(chistat) + " p-value: " + '{0:0.3f}'.format(chipval))
```

Yes, these data provide sufficient evidence for us to conclude that alcoholism is more likely to be present in families of subjects with unipolar depression. ($p<0.05$)

a) Construct a 95% confidence interval for the true difference in population.

```python
from statsmodels.stats.weightstats import DescrStatsW
import numpy as np
hrs12 = np.array([73,58,67,93,33,18,147,110,85,79])
hrs24 = np.array([24,27,49,59,0,11,43,55,44,30])
DescrStatsW(hrs12-hrs24).tconfint_mean()
```

b) Please state the null and alternative hypothesis. Do these data provide sufficient evidence to believed that mean cotinine level 12 hours must be higher than the mean cotinine level 24 hours after smoking. Let α = 0.05.

> $H_0 : \mu_{12hrs} - \mu_{24hrs} \leqslant 0$
>
> $H_0 : \mu_{12hrs} - \mu_{24hrs} > 0$

```python
tstat, pval, df = DescrStatsW(hrs12-hrs24).ttest_mean(alternative="larger")
print("tstat: " + "{0:0.3f}".format(tstat) + " p-value: " + '{0:0.3f}'.format(pval))
```

> Yes, these data provide sufficient evidence to believed that mean cotinine level 12 hours must be higher than the mean cotinine level 24 hours after smoking. ($p<0.01$)
